```python
#to check correlation/association btw categorical columns
"""Test Statistics

H0: Ctegorical columns are not associated with Approved_Flag

H1: Ctegorical columns are associated with Approved_Flag
"""
```

```python
#Chi_square_test
for i in ['MARITALSTATUS','EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']:
  chi2, pval, dof, expected = chi2_contingency(pd.crosstab(df[i],df['Approved_Flag']))
  print(i,'----', pval)
```

```python
"""Since all the p-values are less than 0.05 which implies all columns are accepted."""
```

```python
#VIF for numerical coloumns
numeric_columns=[]
for i in df.columns:
  if df[i].dtype!='object' and i not in ['PROSPECTID','Approved_Flag']:
    numeric_columns.append(i)
    print(i)
```

```python
len(numeric_columns)
```

```python
vif_data = df[numeric_columns]
total_columns = vif_data.shape[1]
columns_to_be_kept = []
column_index = 0


for i in range (0,total_columns):

    vif_value = variance_inflation_factor(vif_data, column_index)
    print (column_index,'---',vif_value)


    if vif_value <= 6:
        columns_to_be_kept.append( numeric_columns[i] )
        column_index = column_index+1

    else:
        vif_data = vif_data.drop([ numeric_columns[i] ] , axis=1)

len(columns_to_be_kept)
```

```python
len(columns_to_be_kept)
```

```python
# Check Anova for columns_to_be_kept

from scipy.stats import f_oneway

columns_to_be_kept_numerical=[]

for i in columns_to_be_kept:
  a= list(df[i])
  b= list(df['Approved_Flag'])

  group_P1=[value for value, group in zip(a,b) if group =='P1']
  group_P2=[value for value, group in zip(a,b) if group =='P2']
  group_P3=[value for value, group in zip(a,b) if group =='P3']
  group_P4=[value for value, group in zip(a,b) if group =='P4']

  f_statistics, p_value =f_oneway(group_P1, group_P2, group_P3,group_P4)

  if p_value<0.05:
    columns_to_be_kept_numerical.append(i)
    print(i)

len(columns_to_be_kept_numerical)
```

```python
# listing all the final features
features = columns_to_be_kept_numerical + ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']
df = df[features + ['Approved_Flag']]
```

# Model Fitting

```python
### 1. Decision Trees
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_classifier=DecisionTreeClassifier(max_depth=20, min_samples_split=10)
```

```python
y = df_encoded['Approved_Flag']
x = df_encoded. drop(['Approved_Flag'], axis = 1 )
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
dt_classifier.fit(x_train, y_train)
```

```python
y_pred = dt_classifier.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print (f"The accuracy_score is:", accuracy)
```

```python
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred)

for i, v in enumerate(['p1', 'p2', 'p3', 'p4']):
    print(f"Class {v}:")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
```
